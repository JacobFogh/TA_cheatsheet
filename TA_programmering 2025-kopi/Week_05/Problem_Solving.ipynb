### Problem 5.7: Parts to Ratio

```python
def parts_to_ratio(part, other_part):
    part_cal = part / (part + other_part)
    return f"The ratio of part in the total mixture is {part_cal}"

print(parts_to_ratio(1,4))
```

### Problem 5.8: Dilute Solution

```python
def dilute(concentation, part_solution, part_solvent):
    cal = (part_solution / (part_solution + part_solvent)) * concentation
    return f"The concentration after dilution is {cal}"

print(dilute(0.5, 2, 3))
```

### Problem 5.9: Profit Margin

```python
def profit_margin(cost, revenue):
    margin = (revenue - cost) / revenue
    return margin * 100

print(profit_margin(20, 40))

print(profit_margin(30*10**9, 31*10**9))

print(profit_margin(200000, 144000))

print(profit_margin(60*10**9, 83*10**9))
```

### Problem 5.10: Beam Deflection

```python
def calculate_beam_deflection(length, load, modulus_of_elasticity, moment_of_inertia):
    D = (length * load)**3 / (3 * modulus_of_elasticity * moment_of_inertia) 
    return D

print(calculate_beam_deflection(1, 2, 3, 4))
```

### Problem 5.12: Bacterial Growth

```python
def bacterial_growth(initial, growth_rate, max_bact):
    hours = 0
    while True:
        initial += growth_rate * initial * ((max_bact - initial) / max_bact)

        hours += 1
        if initial >= max_bact * 0.9:
            return hours
        if hours >= 7 * 24:
            return -1

bacterial_growth(100, 0.1, 1000)
```

### Problem 5.13: Which Fibonacci?

```python
def fibonacci_check(n):

    første_tal = 0
    andet_tal = 1
    fib = 0

    count = 2
    while True:
        count += 1
        fib = første_tal + andet_tal
        første_tal = andet_tal
        andet_tal = fib
        if fib == n:
            return count
        elif fib > n:
            return -1

print(fibonacci_check(2))
#6
print(fibonacci_check(14))
#-1
print(fibonacci_check(14930352))
#37
```

### Problem 5.14: Normal Range

```python
import math

def normal_range(height):
    upper = math.floor(25 * height**2)
    lower = math.ceil(18.5 * height**2)

    return f"Normal weight range: {lower} to {upper} kg"
```

```python

returned = normal_range(1.73)
expected = 'Normal weight range: 56 to 74 kg'

if returned != expected:
    print('Test for Normal Range failed because returned was:')
    print(repr(returned))
    print('instead of:')
    print(repr(expected))
else:
    print('Test for Normal Range passed')
```

### Problem 5.15: Lucas Number

```python
def lucas_number(i):
    start_number = 2
    second_number = 1

    while True:

        if i == 0:
            return start_number
        elif i == 1:
            return second_number
        else:
            return lucas_number(i - 1) + lucas_number(i - 2)

print(lucas_number(7))
        
```

```python
returned = lucas_number(3)
expected = 4

if returned != expected:
    print('Test for Lucas Number failed because returned was:')
    print(repr(returned))
    print('instead of:')
    print(repr(expected))
else:
    print('Test for Lucas Number passed')
```
